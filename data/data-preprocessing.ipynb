```python
import pandas as pd
import random
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('IS Dataset.xlsx')
df = df[['Age','AY', 'Y9']]
df
```

```python
map_dict = {'1st year': 1, '2nd year': 2, '3rd year': 3, '4th year': 4, 'Fifth year':5}
df.replace(map_dict, inplace=True)
df['Y9'] = df['Y9'].astype('Float64')
df
```

```python
for i, row in df.iterrows():
    if row[0] == '18-21':
        df.iloc[i, 0] = random.randint(18, 21)
    elif row[0] == '22-25':
        df.iloc[i, 0] = random.randint(22, 25)
    
    for j in range(int(row[2])):
        df.iloc[i, 2] += - random.randint(0, 1) * 0.25 - random.randint(0, 1) * 0.25

df
```

```python
df.to_csv('dataset.csv', index=False)
```
